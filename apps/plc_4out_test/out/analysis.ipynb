```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
import csv
import sys
```

```python
ctrlfile = "control_data.txt"
expfile = "update_data.txt"
```

```python
ctrldata = []
temp = []

with open(ctrlfile) as textfile:
    line = True
    while(line):
        line = textfile.readline()
        if (line[:3] == "OUT"):
            temp += [float(line[15:-3])]
        if (line[:4] == "OUT3"):
            ctrldata += [temp]
            temp = []

# print(len(ctrldata))
cdata = np.array(ctrldata).T
```

```python
expdata = []
temp = []

with open(expfile) as textfile:
    line = True
    while(line):
        line = textfile.readline()
        if (line[:3] == "OUT"):
            temp += [float(line[15:-3])]
        if (line[:4] == "OUT3"):
            expdata += [temp]
            temp = []

# print(len(expdata))
edata = np.array(expdata).T
```

```python
cdata0 = cdata[0]
cdata1 = cdata[1]
cdata2 = cdata[2]
cdata3 = cdata[3]

edata0 = edata[0]
edata1 = edata[1]
edata2 = edata[2]
edata3 = edata[3]
```

```python
# Building Histogram

fig, ax = plt.subplots(figsize=(9, 6))
ax.hist([cdata0, edata0], color='bg', label=["Control", "Experimental"])

# ax.stem(dbins1[:-1], dhist1, 'b', markerfmt='bo', label="Control")
# ax.stem(dbins2[:-1], dhist2, 'g', markerfmt='go', label="Experimental")
ax.set_title("Distribution of Response Times (Combined)")
ax.set_xlabel("Response Times (ms)")
# ax.set_xlim(left=-1, right=11)
ax.set_ylabel("Count")
# ax.set_ylim(bottom=0, top=1)

ax.legend(bbox_to_anchor=(1.27,1))
fig.show()

print("Control:", np.average(cdata0))
print("Experiment:", np.average(edata0))
```
